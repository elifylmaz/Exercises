```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #grafiklerin gösterilmesini kolaylaştırır ve plt.show() yazmanı azaltır
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"pink") # g green demek yeşil çizgi 
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa göre Değişim Grafiği")
# yukarıda inline yazdığımız için plt.show() yazmamıza gerek kalmadı
```

```python
numpyDizisi1 = np.linspace(0,100,30)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3 
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-") # yıldızlı gösterir
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"g--") #kesik çizgi halinde gösterir
```

```python
plt.subplot(1,2,1) # 1 sıra olacak, 2 kolon olacak, şuanda 1. grafiği çiziyor
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şuanda 2. grafiği çiziyor
plt.plot(numpyDizisi2,numpyDizisi1,"y--")
```

```python
benimFigure = plt.figure() #boş kanvas 
```

```python
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
```

```python
benimFigure = plt.figure() #boş kanvas 
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) #eksenlerin başlangıç ve bitiş noktaları
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("Grafik başlığı")
```

### 0.2: Sol alt köşenin x koordinatı (figürün genişliğine göre oransal olarak).
### 0.2: Sol alt köşenin y koordinatı (figürün yüksekliğine göre oransal olarak).
### 0.9: Eksenin genişliği (figürün genişliğine göre oransal olarak).
### 0.9: Eksenin yüksekliği (figürün yüksekliğine göre oransal olarak).

# iç içe geçmiş grafikler

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3]) #burdaki sayıları değiştirerek küçük grafiğin konumunu değiştirebilirisn
# 02,05 nereden başlıcağını ; 03,09 da boyutları veriyor

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("(eksen1) Ana grafik")


eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("(eksen2) minik grafik")


```

```python
(benimFigur, benimEksenler) = plt.subplots(1,2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("x ekseni")
    
plt.tight_layout() #görüntüyü düzeltiyor
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "Numpydizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "Numpydizisi ** 3")
#legend açıklama kutusu
yeniEksen.legend(loc = 1) #loc değerlerini değiştiriren yeri değişir 1,2,3,4..
```

```python
yeniFigur = plt.figure(figsize=((6,6))) #eksenlerde boyutlandırıyoruz zaten aam istersen buradan da yapabilirisin

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "Numpydizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "Numpydizisi ** 3")
yeniEksen.legend(loc = 1) #loc değerlerini değiştiriren yeri değişir 1,2,3,4..
```

```python
yeniFigur = plt.figure(dpi=200) #dpi kalitesini belirliyor

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "Numpydizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "Numpydizisi ** 3")
yeniEksen.legend(loc = 1) #loc değerlerini değiştiriren yeri değişir 1,2,3,4..
yeniFigur.savefig("yenifigur.png",dpi=200)
```
